#### Question 15 : Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.

Show sorted dataframe on Win%

```python
ipl_details = (ipl["Team1"].value_counts() + ipl["Team2"].value_counts()).sort_values(ascending=False).reset_index()
```

```python
ipl_details.rename(columns={"count" : "Matches Played", "index" : "Teams"},inplace=True)
```

```python
# ipl_details.drop(columns={"Winning %"},inplace=True)
```

```python
Total_Matches = ipl_details["Matches Played"].sum()
```

```python
ipl_details["Winning %"] = (ipl_details["Matches Played"] / Total_Matches) * 100
```

```python
teams = ipl_details["Teams"].unique()
hwinslist = []
awinslist = []
for team in teams:
    home_wins = ipl[(ipl["Team1"] == team) & (ipl["WinningTeam"] == team)].shape[0]
    away_wins = ipl[(ipl["Team2"] == team) & (ipl["WinningTeam"] == team)].shape[0]
    total_wins = ipl[ipl["WinningTeam"] == team].shape[0]
    home_winsp =( home_wins / total_wins) * 100
    away_winsp = (away_wins / total_wins) * 100
    hwinslist.append(home_winsp)
    awinslist.append(away_winsp)

ipl_details["Home Wins %"] = pd.Series(hwinslist)
ipl_details["Away Wins %"] = pd.Series(awinslist)
```

##### Now to do the first question, I will have to create a custom function that would do the z normalisation on all the columns. This function would be used using the apply function on the groupby object.

```python
def z_normalisation(teams, cols):
    for col in cols:
        std = teams[cols].std()
        mean = teams[cols].mean
        teams[f'{col}_z_norm'] = (teams[col] - mean) / std
        return teams


cols = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"] 
teams = fifa.groupby("Team")
teams.apply(z_normalisation, cols = cols)
```
