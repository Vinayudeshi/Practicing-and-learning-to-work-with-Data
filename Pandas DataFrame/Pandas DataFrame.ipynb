### Filtering a DataFrame.

```python
ipl.head(2)
```

#### Question 1: Find the name of the IPL winning team for all years. 

```python
ipl[~ipl['Season'].duplicated()][["Season",'WinningTeam']]
```

#### Question 2: How many IPL matches finished with a super over?

```python
ipl[ipl["SuperOver"] == 'Y'].shape[0]
```

#### Question 3: How many matches has CSK won in Kolkata?

```python
ipl.head(3)
mask1 = ipl["City"] == "Kolkata"
mask2 = ipl["WinningTeam"] == "Chennai Super Kings"

ipl[mask1 & mask2].shape[0]
```

#### Question 4: Find the percentage Toss winning teams that won the match

```python
(ipl[ipl["TossWinner"] == ipl['WinningTeam']].shape[0]) / ipl['TossWinner'].shape[0] 
```

#### Question 5: How many movies have rating higher than 8 and 10000+ votes.

```python
movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"] > 10000)].shape[0]
```

#### Question 6 : How many action movies are rated > 7.5

```python
movies.head(2)
```
